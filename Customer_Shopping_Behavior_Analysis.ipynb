### Importing pandas and extracting data from csv

```python
import pandas as pd
df = pd.read_csv("customer_shopping_behavior.csv")
```

### Check missing values

```python
df.isnull().sum()
```

```python
# here review rating is only column with null values , we can replace null values with median we got for each category
df['Review Rating'] = df.groupby("Category")['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

### Changing column names to lower case with underscore

```python
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(" ","_")
df= df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

### Doing some feature engineering

```python
# create a column age_group , 4 groups for age
labels = ['Young Adult','Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q =4, labels=labels)
```

```python
df[['age','age_group']].head(10)
```

```python
df['frequency_of_purchases'].unique()
```

```python
# create purchase_frequency_days column , we have column frequency_of_purchases but it is text so we will convert it into numbers as it is easier for analysis
frequency_mapping = {
    'Weekly': 7,
    'Bi-Weekly': 14,
    'Fortnightly': 14,
    'Monthly': 30,
    'Every 3 Months': 90,
    'Quarterly': 90,
    'Annually': 365
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
# since both columns have same values we don't need the 2 of them , drop on of them
```

```python
df.drop(columns =['promo_code_used'],inplace=True)
```
